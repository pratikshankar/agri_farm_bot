```python
mock_df=pd.DataFrame(mock_df)
mock_df_1=pd.DataFrame(mock_df_1)
mock_df=pd.concat([mock_df, mock_df_1], ignore_index=True)
```

```python
disease_names = ['Mastitis', 'Lameness', 'Ketosis', 'Milk Fever', 'Bloat', 'Acidosis', 'Brucellosis', 'Johne\'s Disease']
health_status = ['Healthy', 'Sick']
estrus_status = ['In Estrus', 'Not in Estrus']
lactation_number = [1, 2, 3, 4, 5]
feed_type = ['Grass', 'Silage', 'Hay', 'Concentrate', 'Total Mixed Ration (TMR)']
for i in range(len(mock_df)):
    if random.random() < 0.1:  # 10% chance of being sick
        mock_df.at[i, 'health_status'] = random.choice(health_status[1:])
        mock_df.at[i, 'disease_name'] = random.choice(disease_names)
    else:
        mock_df.at[i, 'health_status'] = health_status[0]
        mock_df.at[i, 'disease_name'] = None

    if random.random() < 0.5:  # 50% chance of being in estrus
        mock_df.at[i, 'estrus_status'] = random.choice(estrus_status)
    else:
        mock_df.at[i, 'estrus_status'] = estrus_status[1]

    mock_df.at[i, 'feed_type'] = random.choice(feed_type)
```

```python
mock_df=mock_df.drop(columns=['lactation_number'])
```

```python
mock_df.to_csv('mock_data_farm_2 years.csv', index=False)
```
